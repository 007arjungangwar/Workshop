Create the function who will take stadium name And run as input from the user and Return top_10_batsman_run_in_stadium (That particular stadium)

```python
def top_10_batsman_run_in_stadium(stadium,run):
    stadium=df3.groupby("venue").get_group(stadium)
    mask=(stadium["batsman_runs"]==run)
    stadium[mask]
    return stadium[mask]["batter"].value_counts().head(10)

top_10_batsman_run_in_stadium("Wankhede Stadium",6)
```

```python
venue_list=df3["venue"].value_counts().index.to_list()
venue_list
```

can you give top_10_batsman_run_in_stadium (each stadium)

```python
for i in venue_list:
    print("venue",i,end="\n")
    print(f"{top_10_batsman_run_in_stadium(i,6)}")
```

Can you create a function so that we can get Idea on particular stadium top number of Batsman make perticular runs.

```python
def top_no_batsman_run_in_stadium(stadium,run,no):
    stadium=df3.groupby("venue").get_group(stadium)
    mask=(stadium["batsman_runs"]==run)
    stadium[mask]
    return stadium[mask]["batter"].value_counts().head(no)

top_no_batsman_run_in_stadium("Wankhede Stadium",6,2)
```

```python
for i in venue_list:
    print("venue",i,end="\n")
    print(f"{top_no_batsman_run_in_stadium(i,4,2)}")
```
